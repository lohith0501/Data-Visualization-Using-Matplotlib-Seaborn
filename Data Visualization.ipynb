```python
plt.scatter(y,z)
plt.title("Sin(x) vs Cos(x)")
plt.xlabel("Sinx")
plt.ylabel("Cosx")
```

# Data Visualization Using Seaborn

```python
import seaborn as sns
```

```python
tips=sns.load_dataset("tips")
titanic=sns.load_dataset("titanic")
```

## Count plot

```python
sns.countplot(x="sex",hue="class",data=titanic)
```

## Strip Chart

```python
sns.stripplot(x="sex",y="total_bill",hue="smoker",data=tips)
```

```python
sns.swarmplot(x="sex",y="total_bill",hue="smoker",data=tips)
```

```python
sns.violinplot(x="sex",y="total_bill",hue="smoker",data=tips)
```

##  Barplot with statistics

```python
sns.barplot(x="sex",y="age",hue="class",data=titanic,order=["male","female"],estimator=np.mean)
```

## Boxplot

```python
sns.boxplot(x="sex",y="age",hue="class",data=titanic)
```

## Rel plots

```python
sns.relplot(x="age",y="fare",hue="pclass",row="survived",col="sex",kind="scatter",data=titanic,palette="coolwarm")
```

## Regression plots

```python
sns.lmplot(x="total_bill",y="tip",hue="smoker",col="day",data=tips,
          markers=[0,4],col_wrap=3)
```

## Heatmap

```python
sns.heatmap(titanic.corr())
```

# BITS Class Work Basic Matplotlib

```python
x=np.linspace(0,5,10)
y=np.linspace(3,6,10)
```

```python
plt.plot(x,y,"ro",label="x and y")
plt.plot(x,y**2,"g+",label="Square of y")
plt.plot(x,y**3,"yo",label="Cube of y")
plt.legend(frameon=False)
plt.grid(True)
plt.xlabel("Value of X")
plt.ylabel("Values of Y")
plt.title("Variation of Y over X")
```

# BITS Class Work Advanced Matplotlib 

```python
x=np.linspace(-np.pi,np.pi,256,endpoint=True)
C,S=np.cos(x),np.sin(x)
```

```python
plt.plot(x,C)
plt.plot(x,S)
```

## Spines to remove borders

```python
plt.plot(x,C)
plt.plot(x,S)
ax=plt.gca()
# ax.spines['right'].set_color('none') # Change color or remove spines
ax.spines['top'].set_visible(False) # Better approach to remove spines

ax.spines['bottom'].set_position(('data',0))
ax.xaxis.set_ticks_position('bottom')
ax.spines['left'].set_position(('data',0))
ax.yaxis.set_ticks_position('left')
```

## Dual Axis Chart

```python
fig=plt.figure()
plt1=fig.add_subplot(1,2,1)

plt1.plot(x,C)

plt2=plt1.twinx()
plt2.plot(x,S,'b')
```

## Log Application

```python
dt=pd.read_csv("market_fact.csv")
```

```python
dt.head()
```

```python
plt.subplot(1,2,1)
plt.boxplot(dt["Sales"])
plt.show()

plt.subplot(1,2,2)
plt.boxplot(dt["Sales"])
plt.yscale("log")

plt.tight_layout()
```

## Plot using groupby & aggregate

```python
new_data=dt.groupby(["Prod_id"])["Sales"].aggregate(np.sum).reset_index().sort_values("Sales",ascending=False)
```

```python
plt.bar("Prod_id","Sales",data=new_data)
plt.xticks(rotation=70)
plt.show()
```

## Joint plot using seaborn

```python
sns.jointplot(dt.Sales,dt.Profit,dt)
```

## Dist plot using seaborn with rug 

```python
sns.distplot(dt["Profit"],rug=True)
```

## Joint plot using condition

```python
dt_filter=dt[(dt.Profit>-100) & (dt.Profit<100) & (dt.Sales<200)]
```

```python
sns.jointplot(dt_filter.Sales,dt_filter.Profit,dt_filter,kind="hex")
```

```python
sns.jointplot(dt_filter.Sales,dt_filter.Profit,dt_filter,kind="kde")
```
